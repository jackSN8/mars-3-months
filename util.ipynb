```python

def clear_csv_file(csv_filename):
    # Open the file in write mode, which will erase its content.
    with open(csv_filename, 'w') as f:
        # Optionally, call truncate() to ensure the file is cleared.
        f.truncate()
    print(f"Cleared the file: {csv_filename}")

csv_file = "accelerations_EM.csv"
```

```python


# Load your CSV file; change 'data.csv' to your file path
df = pd.read_csv('accelerations_ME.csv')

# Let the user specify the length of the x-axis
x_length = 300

# Generate the x-axis; here we create a linearly spaced array
# You can adjust the start, stop, or use np.arange(x_length) as needed.
x = np.linspace(0, 10, x_length)

# Create a subplot for each column, stacked vertically
num_plots = len(df.columns)
fig, axes = plt.subplots(num_plots, 1, figsize=(10, 2 * num_plots), sharex=True)

# If there is only one column, axes is not an array, so we wrap it in a list
if num_plots == 1:
    axes = [axes]

# Plot each column against the generated x-axis
for ax, col in zip(axes, df.columns):
    # Get the data for the column; if the CSV column has more or fewer rows than x_length,
    # you may need to adjust how you handle it (e.g., slicing, interpolation, etc.)
    data = df[col].values[:x_length]  # here we slice to match the x-axis length
    ax.plot(x, data, marker='o', linestyle='-')
    ax.set_title(col)
    ax.grid(True)

# Set a common label for the x-axis
plt.xlabel('X Axis')
plt.tight_layout()
plt.show()
```
